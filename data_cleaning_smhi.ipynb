```python
import pandas as pd

```

```python
weather_data_smhi= pd.read_json('data/weather_output_smhi.json')
weather_data_smhi_stockholm= weather_data_smhi[weather_data_smhi['city'] == 'Stockholm']

weather_data_smhi_stockholm
```

```python
weather_data_smhi_stockholm = weather_data_smhi_stockholm.iloc[0]
weather_data_smhi_stockholm
```

```python
wd_stockholm_smhi_dict=weather_data_smhi_stockholm['weather_data']
timeseries_stockholm_smhi=wd_stockholm_smhi_dict['timeSeries']

timeseries_stockholm_smhi
```

```python
smhi_stockholm_df=pd.json_normalize(timeseries_stockholm_smhi)
smhi_stockholm_df
```

```python
exploded = smhi_stockholm_df.explode('parameters')
smhi_stockholm_df_mod = pd.concat([exploded['validTime'].reset_index(drop=True), pd.json_normalize(exploded['parameters'])], axis=1)

smhi_stockholm_df_mod['values'] = smhi_stockholm_df_mod['values'].str[0]
```

```python
#smhi_oslo_df_mod.pivot(index=smhi_oslo_df.index, columns='name')['values']

# smhi_oslo_df_mod.reset_index().groupby(['validTime', 'name'])['values'].aggregate('first').unstack()

smhi_stockholm_df_pivot = smhi_stockholm_df_mod.pivot(index='validTime', columns='name', values='values').reset_index()
smhi_stockholm_df_pivot = smhi_stockholm_df_pivot.head(24)
```

```python
smhi_stockholm_df_pivot.drop(columns=['gust', 'lcc_mean', 'mcc_mean', 'hcc_mean', 'spp', 'pcat', 'pmin', 'pmax', 'pmedian', 'wsymb2', 'tstm'])
```

```python
smhi_stockholm_df_pivot.to_csv('data/test.csv', index=False)
```
